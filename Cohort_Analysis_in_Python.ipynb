## Menampilkan data

```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Salinan Salinan Salinan Online Retail Data.csv', engine='python', encoding= 'ISO-8859-1')
data.head()

#encoding = 'ISO-8859-1' maksudnya adalah didalam file terdapat huruf yang tidak biasa ex:bhspanyol sehingga agar python bisa membacanya maka menggunakan ini
```

## Reformat Timestamp

```python
import dateutil
from datetime import datetime as dt
from pytz import utc

#membuat kolom baru 'datetime' 'month' 'year'
data['datetime'] = data['order_date'].apply(lambda x: dateutil.parser.parse(x).timestamp())
data['month'] = data['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).month)
data['year'] = data['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).year)
```

```python
data.head()
```

## Membuat Cohort

```python
# 'AUG 2010'
# '2010 AUG'
# AUG 2010 -> 201008

data['cohort'] = data.apply(lambda row: (row['year'] * 100) + (row['month']), axis=1)
# 2010 * 100 = 201000
# aug -> 201000 + 08 = 201008

cohorts = data.groupby('customer_id')['cohort'].min().reset_index()
cohorts.columns = ['customer_id', 'first_cohort']

data = data.merge(cohorts, on='customer_id', how='left')
```

```python
cohorts.head()
```

```python
data.head()
```

## Membuat header untuk setiap cohort

```python
headers = data['cohort'].value_counts().reset_index()
headers.columns = ['Cohorts', 'Count']
headers.head()
headers = headers.sort_values(['Cohorts'])['Cohorts'].to_list()
```

```python
headers
```

## Pivot Data berdasarkan Cohort

```python
data.head()
```

```python
#menghapus data nan
data.dropna(inplace=True)
```

```python
#membuat kolom 'cohort distance'
data['cohort_distance'] = data.apply(lambda row: (headers.index(row['cohort']) - headers.index(row['first_cohort'])) if (row['first_cohort'] != 0 and row['cohort'] != 0) else np.nan, axis=1)
```

```python
data.head()
```

```python
cohort_pivot = pd.pivot_table(data,
                              index='first_cohort',
                              columns='cohort_distance',
                              values='customer_id',
                              aggfunc=pd.Series.nunique)
```

```python
cohort_pivot
```

```python
cohort_pivot = cohort_pivot.div(cohort_pivot[0],axis=0)
```

```python
cohort_pivot
```

## Menampilkan ke dalam bentuk heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig_dims = (12, 8)
fig, ax = plt.subplots(figsize=fig_dims)
sns.heatmap(cohort_pivot, annot=True, fmt='.0%', mask=cohort_pivot.isnull(), ax=ax, square=True, linewidths=.5, cmap=sns.cubehelix_palette(8))

plt.show()
```
